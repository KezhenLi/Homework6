```python
from numpy import array, arange, pi, sin, cos, sqrt
from pylab import plot, show, xlabel, ylabel
```

```python
#(a) ax = 0, ay = -g, F = D*v^2, D = 0.5*pi*R^2*rho*C.
#fx = -D*v*vx = m*ax, fy = -D*v*vy -m*g = m*ay ,
#ax = -D*v*vx / m, ay = -g - D*v*vy / m

#(b)
g = 9.81
m = 1  
R = 0.08  # m
theta0 = 30 * pi / 180  
v0 = 100  
rho = 1.22  
C = 0.47  
t0 = 0
t1 = 10
N = 10000
h = (t1 - t0) / N

def f(r, t, m):
    x = r[0]
    vx = r[1]
    y = r[2]
    vy = r[3]
    v = sqrt(vx ** 2 + vy ** 2)
    return array([vx, - pi * R ** 2 * rho * C / (2 * m) * vx * v,
                  vy, -g - pi * R ** 2 * rho * C / (2 * m) * vy * v], float)

tpoints = arange(t0, t1, h)
xpoints1 = []
ypoints1 = []
r = array([0, v0 * cos(theta0), 0, v0 * sin(theta0)], float)
for t in tpoints:
    xpoints1.append(r[0])
    ypoints1.append(r[2])
    k1 = h * f(r, t, m)
    k2 = h * f(r + 0.5 * k1, t + 0.5 * h, m)
    k3 = h * f(r + 0.5 * k2, t + 0.5 * h, m)
    k4 = h * f(r + k3, t + h, m)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(xpoints1, ypoints1)
xlabel("x(m)")
ylabel("y(m)")
show()

#(c)
m = 0.5

def f(r, t, m):
    x = r[0]
    vx = r[1]
    y = r[2]
    vy = r[3]
    v = sqrt(vx ** 2 + vy ** 2)
    return array([vx, - pi * R ** 2 * rho * C / (2 * m) * vx * v,
                  vy, -g - pi * R ** 2 * rho * C / (2 * m) * vy * v], float)

tpoints = arange(t0, t1, h)
xpoints2 = []
ypoints2 = []
r = array([0, v0 * cos(theta0), 0, v0 * sin(theta0)], float)
for t in tpoints:
    xpoints2.append(r[0])
    ypoints2.append(r[2])
    k1 = h * f(r, t, m)
    k2 = h * f(r + 0.5 * k1, t + 0.5 * h, m)
    k3 = h * f(r + 0.5 * k2, t + 0.5 * h, m)
    k4 = h * f(r + k3, t + h, m)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(xpoints2, ypoints2)
xlabel("x(m)")
ylabel("y(m)")
show()

m = 1.5

def f(r, t, m):
    x = r[0]
    vx = r[1]
    y = r[2]
    vy = r[3]
    v = sqrt(vx ** 2 + vy ** 2)
    return array([vx, - pi * R ** 2 * rho * C / (2 * m) * vx * v,
                  vy, -g - pi * R ** 2 * rho * C / (2 * m) * vy * v], float)

tpoints = arange(t0, t1, h)
xpoints3 = []
ypoints3 = []
r = array([0, v0 * cos(theta0), 0, v0 * sin(theta0)], float)
for t in tpoints:
    xpoints3.append(r[0])
    ypoints3.append(r[2])
    k1 = h * f(r, t, m)
    k2 = h * f(r + 0.5 * k1, t + 0.5 * h, m)
    k3 = h * f(r + 0.5 * k2, t + 0.5 * h, m)
    k4 = h * f(r + k3, t + h, m)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(xpoints3, ypoints3)
xlabel("x(m)")
ylabel("y(m)")
show()
#The cannonball travels further when it is heavier.
plot(xpoints1, ypoints1)
plot(xpoints2, ypoints2)
plot(xpoints3, ypoints3)
xlabel("x(m)")
ylabel("y(m)")
show()

```
