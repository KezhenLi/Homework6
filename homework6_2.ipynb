```python
from numpy import array, arange
from pylab import plot, show, xlabel, ylabel
```

```python
#(a)
alpha = 1
beta = 0.5
gamma = 0.5
delta = 2
t0 = 0
t1 = 30
N = 10000
h = (t1 - t0) / N

def f(r):
    x = r[0]
    y = r[1]
    fx = alpha * x - beta * x * y
    fy = gamma * x * y - delta * y
    return array([fx, fy] , float)

tpoints = arange(t0, t1, h)
xpoints = []
ypoints = []

r = array([2.0, 2.0], float)
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    k1 = h * f(r)
    k2 = h * f(r + 0.5 * k1)
    k3 = h * f(r + 0.5 * k2)
    k4 = h * f(r + k3)
    r += (k1 + 2 * k2 + 2 * k3 + k4)

plot(tpoints, xpoints)
plot(tpoints, ypoints)
xlabel("t")
ylabel("x(t), y(t)")
show()
#(b)The rabbit population and fox population are increased oscillations, and the fox population oscillates later than the rabbit population.
```
