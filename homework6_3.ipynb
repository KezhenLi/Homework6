```python
from numpy import array, arange
from pylab import plot, show, xlabel, ylabel
```

```python
#(a)
omega = 1
t0 = 0
t1 = 50
N = 1000  
h = (t1 - t0) / N

def harmonic(r, t):
    x = r[0]
    v = r[1]
    return array([v, - omega ** 2 * x], float)

tpoints = arange(t0, t1, h)
xpoints = []
r = array([1.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    k1 = h * harmonic(r, t)
    k2 = h * harmonic(r + 0.5 * k1, t + 0.5 * h)
    k3 = h * harmonic(r + 0.5 * k2, t + 0.5 * h)
    k4 = h * harmonic(r + k3, t + h)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(tpoints, xpoints)
xlabel("t(s)")
ylabel("x(m)")
show()

#(b)
tpoints = arange(t0, t1, h)
xpoints = []
r = array([2.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    k1 = h * harmonic(r, t)
    k2 = h * harmonic(r + 0.5 * k1, t + 0.5 * h)
    k3 = h * harmonic(r + 0.5 * k2, t + 0.5 * h)
    k4 = h * harmonic(r + k3, t + h)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(tpoints, xpoints)
xlabel("t(s)")
ylabel("x(m)")
show()

#(c)
def anharmonic(r, t):
    x = r[0]
    v = r[1]
    return array([v, - omega ** 2 * x ** 3], float)

xpoints = []
vpoints = []
r = array([1.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    vpoints.append(r[1])
    k1 = h * f_anharmonic(r, t)
    k2 = h * f_anharmonic(r + 0.5 * k1, t + 0.5 * h)
    k3 = h * f_anharmonic(r + 0.5 * k2, t + 0.5 * h)
    k4 = h * f_anharmonic(r + k3, t + h)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(tpoints, xpoints)
xlabel("t(s)")
ylabel("x(m)")
show()

xpoints = []
vpoints = []
r = array([2.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    vpoints.append(r[1])
    k1 = h * f_anharmonic(r, t)
    k2 = h * f_anharmonic(r + 0.5 * k1, t + 0.5 * h)
    k3 = h * f_anharmonic(r + 0.5 * k2, t + 0.5 * h)
    k4 = h * f_anharmonic(r + k3, t + h)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(tpoints, xpoints)
xlabel("t(s)")
ylabel("x(m)")
show()

#(d)
plot(xpoints, vpoints)
xlabel("x")
ylabel("v")
show()

#(e)
t1 = 20
N = 10000  
h = (t1 - t0) / N
mu = 1
def g(r, t, mu):
    x = r[0]
    v = r[1]
    return array([v, mu * (1 - x ** 2) * v - omega ** 2 * x], float)

tpoints = arange(t0, t1, h)

xpoints = []
vpoints = []
r = array([1.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    vpoints.append(r[1])
    k1 = h * g(r, t, mu)
    k2 = h * g(r + 0.5 * k1, t + 0.5 * h, mu)
    k3 = h * g(r + 0.5 * k2, t + 0.5 * h, mu)
    k4 = h * g(r + k3, t + h, mu)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(xpoints, vpoints)
xlabel("x")
ylabel("v")
show()

t1 = 20
N = 10000  
h = (t1 - t0) / N
mu = 2
def g(r, t, mu):
    x = r[0]
    v = r[1]
    return array([v, mu * (1 - x ** 2) * v - omega ** 2 * x], float)

tpoints = arange(t0, t1, h)

xpoints = []
vpoints = []
r = array([1.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    vpoints.append(r[1])
    k1 = h * g(r, t, mu)
    k2 = h * g(r + 0.5 * k1, t + 0.5 * h, mu)
    k3 = h * g(r + 0.5 * k2, t + 0.5 * h, mu)
    k4 = h * g(r + k3, t + h, mu)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(xpoints, vpoints)
xlabel("x")
ylabel("v")
show()

t1 = 20
N = 10000  
h = (t1 - t0) / N
mu = 4
def g(r, t, mu):
    x = r[0]
    v = r[1]
    return array([v, mu * (1 - x ** 2) * v - omega ** 2 * x], float)

tpoints = arange(t0, t1, h)

xpoints = []
vpoints = []
r = array([1.0, 0.0], float)
for t in tpoints:
    xpoints.append(r[0])
    vpoints.append(r[1])
    k1 = h * g(r, t, mu)
    k2 = h * g(r + 0.5 * k1, t + 0.5 * h, mu)
    k3 = h * g(r + 0.5 * k2, t + 0.5 * h, mu)
    k4 = h * g(r + k3, t + h, mu)
    r += (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(xpoints, vpoints)
xlabel("x")
ylabel("v")
show()
```
