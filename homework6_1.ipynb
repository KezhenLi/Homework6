```python
from numpy import floor, arange
from pylab import plot, show, xlabel, ylabel
```

```python
#(a)
t0 = 0
t1 = 10
N = 1000
h = (t1 - t0) / N  
RC1 = 0.01
RC2 = 0.1
RC3 = 1

def Vin(t):
    if floor(2 * t) % 2 == 0:
        return 1
    else:
        return -1

def f(V, t, RC):
    return 1 / RC * (Vin(t) - V)

tpoints = arange(t0, t1, h)
Vpoints1 = []
Vpoints2 = []
Vpoints3 = []
V1 = 0
V2 = 0
V3 = 0

for t in tpoints:
    Vpoints1.append(V1)
    k1 = h * f(V1, t, RC1)
    k2 = h * f(V1 + 0.5 * k1, t + 0.5 * h, RC1)
    k3 = h * f(V1 + 0.5 * k2, t + 0.5 * h, RC1)
    k4 = h * f(V1 + k3, t + h, RC1)
    V1 +=  (k1 + 2 * k2 + 2 * k3 + k4) / 6
    Vpoints2.append(V2)
    k1 = h * f(V2, t, RC2)
    k2 = h * f(V2 + 0.5 * k1, t + 0.5 * h, RC2)
    k3 = h * f(V2 + 0.5 * k2, t + 0.5 * h, RC2)
    k4 = h * f(V2 + k3, t + h, RC2)
    V2 +=  (k1 + 2 * k2 + 2 * k3 + k4) / 6
    Vpoints3.append(V3)
    k1 = h * f(V3, t, RC3)
    k2 = h * f(V3 + 0.5 * k1, t + 0.5 * h, RC3)
    k3 = h * f(V3 + 0.5 * k2, t + 0.5 * h, RC3)
    k4 = h * f(V3 + k3, t + h, RC3)
    V3 +=  (k1 + 2 * k2 + 2 * k3 + k4) / 6

plot(tpoints, Vpoints1)
plot(tpoints, Vpoints2)
plot(tpoints, Vpoints3)
xlabel('t')
ylabel('V(t)')
show()
# h = (10 - 0) / 1000 = 0.01 
#(b)When RC = 0.01, Vout is almost same as Vin; When RC = 0.1, Vout is closed to Vin; When RC = 1, it's hard to see the shape of Vin.
#Low-pass filter passes signals with a frequency lower than a selected cutoff frequency(given by 1 / (2*pi*RC)) and blocks signals with frequencies higher than the cutoff frequency. 
```
